# **`let's visualise the data`**

```python
sns.pairplot(data)
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(20,10))
sns.countplot(data['age'] , data =data)
```

# **`From the above graph we can conclude that :`**

***`->> The age of people who mostly apply for the credit card  = 50-60 year old`***

***`->> Mostly people of  56 yr old are interested in applying  for the credit card`***

***`->> Mostly People  of 39 yr old are not interested in applying for the credit card`***

```python
data['gender'].value_counts()
```

```python
fig1, ax1 = plt.subplots(figsize=(6, 5))

gender1 =data.gender.value_counts()
labels=np.array(gender1.index)
sizes = gender1.values
colors = ['gold', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)

plt.title('Gender Comparison ')
plt.axis('equal')
plt.show()
```

# ***From the above graph we can conclude that :***

---



***`->> Female are more active in applying crerdit card in comparison to male`***

```python
data['decision'].value_counts()
```

```python
fig1, ax1 = plt.subplots(figsize=(6, 5))

approval =data.decision.value_counts()
labels=np.array(approval.index)
sizes = approval.values
colors = ['gold', 'lightskyblue','red']

plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)

plt.title('Approval Result')
plt.axis('equal')
plt.show()
```

# ***`From the above pie chart we can conclude that `***
---
***`chance of approval of credit card applicatio is 69.9%`***

***`chance of pending of credit card application is 8.8%`***

***`chance of decline / reject of credit card applicatio is 21.3%`***
